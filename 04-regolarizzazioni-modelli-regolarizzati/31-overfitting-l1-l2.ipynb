```python
import pandas as pd
```

```python
real_estate = pd.read_csv('../00-data/Real estate.csv')
real_estate.columns=['ID', 'transaction_date', 'house_age', 'nearest_MRT_station_distance', 'minimarket_nr', 'lat', 'long', 'price_m2']

real_estate
```

```python
north_east_point = (real_estate['lat'].max(), real_estate['long'].max())
north_east_point
```

```python
import numpy as np

real_estate_ne = real_estate.copy()
real_estate_ne['distance_from_NE'] = np.sqrt(
    (north_east_point[0] - real_estate_ne['lat'])**2 +
    (north_east_point[1] - real_estate_ne['long'])**2
)

real_estate_ne
```

```python
output = ['price_m2']
X = real_estate.drop(output, axis=1).values
y = real_estate[output].values
```

```python
print(X)
print(y)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.3)

print(f"{X_train.shape}, {y_train.shape}, {X_test.shape}, {y_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
```

```python
ss = StandardScaler()
X_train_std = ss.fit_transform(X_train)
X_test_std = ss.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
scores = []

for i in range(1,11):
    pf = PolynomialFeatures(degree=i)
    X_train_poly = pf.fit_transform(X_train_std)
    X_test_poly = pf.fit_transform(X_test_std)
    ll = LinearRegression()
    ll.fit(X_train_poly, y_train)
    y_pred_train = ll.predict(X_train_poly)
    y_pred_test = ll.predict(X_test_poly)
    print(f'poly: {i}')
    print(f'TRAIN -> mse: {mean_squared_error(y_train, y_pred_train)}; r2: {r2_score(y_train, y_pred_train)}')
    print(f'TEST -> mse: {mean_squared_error(y_test, y_pred_test)}; r2: {r2_score(y_test, y_pred_test)}')
    print('-----------------------------------------------------')
```

```python
from sklearn.linear_model import Ridge, Lasso, ElasticNet
```

```python
alphas = [0.001, 0.01, 0.1, 1, 10]
polys = [2,3,4]
```

```python
for poly in polys:
  print(f'poly: {poly}')
  pf = PolynomialFeatures(degree=poly)
  X_train_poly = pf.fit_transform(X_train_std)
  X_test_poly = pf.transform(X_test_std)
  
  for alpha in alphas:
    print(f'alpha: {alpha}')
    ridge = Ridge(alpha=alpha)
    ridge.fit(X_train_poly, y_train)
    
    y_pred_train_ridged = ridge.predict(X_train_poly)
    y_pred_test_ridged = ridge.predict(X_test_poly)
    
    print(f'TRAIN -> mse: {mean_squared_error(y_pred_train_ridged, y_train)}; r2: {r2_score(y_pred_train_ridged, y_train)}')
    print(f'TEST -> mse: {mean_squared_error(y_pred_test_ridged, y_test)}; r2: {r2_score(y_pred_test_ridged, y_test)}')
    print('-----------------------------------------------------')
```

```python
l1_ratios = [0.25, 0.5, 0.75]
for l1_ratio in l1_ratios:
  print(f'l1_ratio {l1_ratio}')
  for poly in polys:
    print(f'poly: {poly}')
    pf = PolynomialFeatures(degree=poly)
    X_train_poly = pf.fit_transform(X_train_std)
    X_test_poly = pf.transform(X_test_std)
    
    for alpha in alphas:
      print(f'alpha: {alpha}')
      lassoe = ElasticNet(alpha=alpha, l1_ratio=l1_ratio)
      lassoe.fit(X_train_poly, y_train)
      
      y_pred_train_lassoed = lassoe.predict(X_train_poly)
      y_pred_test_lassoed = lassoe.predict(X_test_poly)
      
      print(f'TRAIN -> mse: {mean_squared_error(y_pred_train_lassoed, y_train)}; r2: {r2_score(y_pred_train_lassoed, y_train)}')
      print(f'TEST -> mse: {mean_squared_error(y_pred_test_lassoed, y_test)}; r2: {r2_score(y_pred_test_lassoed, y_test)}')
      print('-----------------------------------------------------')
```
