```python
from math import gcd, prod
```

```python
def coprimi_wheel(base):
    coprimes = []
    last = 1
    i = 0
    
    while last < base:
        if gcd(base, i) == 1:
            coprimes.append(i)
            last = i
        i += 1 

    return coprimes
```

```python
def calculate_detas(base):
  coprimes = coprimi_wheel(base)
  deltas = []

  for i in range(1, len(coprimes)):
    deltas.append(coprimes[i] - coprimes[i-1])
    
  return deltas
    
```

```python
def calculate_nrs (deltas, init=1):
  res = [init]
  for i in deltas:
    if len(res) > 1 and res[-1] + i == res[1]*res[1]:
      return res
    res.append(res[-1] + i)
  
  return res
```

```python
def rotate_array(arr, n):
    n = n % len(arr)
    return arr[n:] + arr[:n]
```

```python
def calculate_primes(initPrimes, rotate=0, minimum=1):
 seed = prod(initPrimes)
 deltas = calculate_detas(seed)
 return calculate_nrs(rotate_array(deltas, rotate), minimum)
```
