```python
from sklearn.datasets import make_circles
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
```

```python
X, y = make_circles(noise=0.3, n_samples=600, factor=0.5, random_state=42)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y)

print(f'X_train: {X_train.shape}, X_test: {X_test.shape}, y_train: {y_train.shape}, y_test: {y_test.shape}')
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

```python
kernels = ['linear', 'rbf', 'sigmoid', 'poly']
```

```python
for k in kernels:
  print(f'kernel {k}')
  svc = SVC(kernel=k, probability=True)
  svc.fit(X_train, y_train)
  print(f'test score: {svc.score(X_train, y_train)}')
  print(f'test score: {svc.score(X_test, y_test)}')
  print('-'*20)
```

attualmente il kernel rbf da la migliore soluzione, proviamo a giocare con il parametro c

```python
cs = [0.01, 0.1, 1, 10, 100, 1000]
```

```python
for c in cs:
  print(f'c {c}')
  svc = SVC(kernel='rbf', C=c, probability=True)
  svc.fit(X_train, y_train)
  print(f'test score: {svc.score(X_train, y_train)}')
  print(f'test score: {svc.score(X_test, y_test)}')
  print('-'*20)
```

probabilmente la soluzione migliore  con kernel rbf e c 100
