```python
#creo una maschera che filtra tutti i paesi dell'unione europea e cambia il loro nome in "UE"
eu_countries = [
    'France', 'Netherlands', 'Germany', 'Spain', 'Poland', 'Portugal',
    'Italy', 'Belgium', 'Lithuania', 'Denmark', 'Cyprus', 'Sweden',
    'Austria', 'Finland', 'Greece', 'Czech Republic', 'Malta', 'EIRE'
]

df_eu  = df.copy()

europeanMask = df_eu['Country'].isin(eu_countries)

df_eu.loc[europeanMask, 'Country'] = 'UE'
# df_eu_ranamed = df[df['Country'] 

```

```python
#calcolo la percentuale di vendite verso ogni parte considerando l'ue come stato unico
df_eu['TotalSales'] = df_eu['Quantity'] * df_eu['UnitPrice']
sales_per_country = df_eu.groupby('Country')['TotalSales'].sum()
# sales_per_country = df_eu.groupby('Country').apply(lambda x: x['Quantity'] * x['UnitPrice'], axis=1).sum()
```

```python
sales_per_country
```

```python
#guardo la classifica dei piu venduti
sales_per_country.sort_values(ascending=False)
```

```python
percentage_acquired = sales_per_country.apply(lambda x: x/sales_per_country.sum())
percentage_acquired.sort_values(ascending=False)
```

```python
percentage_acquired.info()
```

```python
#verifico che la percentuale è calcolata bene
percentage_acquired.sum()
```

```python
sales_and_percentage = pd.concat([sales_per_country, percentage_acquired], axis=1)
sales_and_percentage.columns=['TotalSales', 'Percentage']

sales_and_percentage
```

proviamo a fare applyMap

```python
sales_and_percentage_remapped = sales_and_percentage.applymap(lambda x: round(x, 3))
sales_and_percentage_remapped
```

annullo tutte le variabili a 0

```python
sales_and_percentage_removed_0 =sales_and_percentage_remapped.copy()
mask_sales_remapped  = sales_and_percentage_removed_0['Percentage'] == 0
# mask_sales_remapped
sales_and_percentage_removed_0.loc[mask_sales_remapped] = float('nan')

sales_and_percentage_removed_0
```

```python
#riempo gli elementi nan con 0

sales_and_percentage_reintegrated_0 = sales_and_percentage_removed_0.copy()

sales_and_percentage_reintegrated_0
```

```python
sales_and_percentage_reintegrated_0.fillna(0, inplace=True)

sales_and_percentage_reintegrated_0
```
