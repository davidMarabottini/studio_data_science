```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../00-data/data.csv', encoding='latin1')

df
```

la normalizzazione e la standardizzazione andranno fatte su quantità (Quantity) e prezzo unitario (UnitPrice)

Procedo con pandas

```python
features = ['Quantity', 'UnitPrice']
to_reduce = df[features]
```

normalizzazione

```python
#normalizzazione, porta i valori in un range tra 0 e 1
df_norm = df.copy()
df_norm[features] = (to_reduce-to_reduce.min())/(to_reduce.max()-to_reduce.min())
df_norm
```

standardizzazione

```python
X_std = df.copy()
X_std[features] = (to_reduce-to_reduce.mean())/(to_reduce.std())
X_std
```

normalizzazione e standardizzazione in array numpy, uso sklearn

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

Normalizzazione (MinMaxScaler)

```python
mms = MinMaxScaler()

X_norm = df.copy()
X_norm[features] = mms.fit_transform(X_norm[features])
X_norm
```

Standardizzazione (StandardScaler)

```python
ss = StandardScaler()

X_std = df.copy()
X_std[features] = ss.fit_transform(X_std[features])
X_std
```
